```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
file_path = "/Users/rajpatil/Machine Learning/Complete-Data-Science-With-Machine-Learning-And-NLP-2024-main/4-Ridge Lasso And Elasticnet/Ridge Lassso Elastic Regression Practicals/Algerian_forest_fires_dataset_UPDATE.csv"
dataset = pd.read_csv(file_path, header = 1)
dataset.head(5)
```

```python
dataset.info()
```

## Data Cleaning

```python
dataset[dataset.isnull().any(axis = 1)]
```

```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:,"Region"] = 1
df = dataset
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.info()
```

```python
df.head(5)
```

## Remove the Null Values

```python
df = df.dropna().reset_index(drop = True)
```

```python
df.isnull().sum()
```

```python
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

## Fix spaces in column names

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
object = [features for features in df.columns if df[features].dtype == 'O']
```

```python
for i in object:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.to_csv('Algerian_forest_cleaned_dataset.csv', index = False)
```
